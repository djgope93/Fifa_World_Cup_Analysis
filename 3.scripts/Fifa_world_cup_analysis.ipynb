##### 1.Top Winning Teams

```python
# Determine the outcome of each match
def determine_outcome(row):
    if row['Home Team Goals'] > row['Away Team Goals']:
        return row['Home Team Name']
    elif row['Home Team Goals'] < row['Away Team Goals']:
        return row['Away Team Name']
    else:
        return 'Draw'

matches_df['Winner'] = matches_df.apply(determine_outcome, axis=1)

# Count the number of wins for each team
win_counts = matches_df[matches_df['Winner'] != 'Draw']['Winner'].value_counts()

# Display the top 10 teams with the most wins
top_winning_teams = win_counts.head(10)
print(top_winning_teams)

matches_df['Outcome'] = matches_df.apply(determine_outcome, axis=1)

# Verify that the Outcome column is created
print(matches_df[['Home Team Name', 'Away Team Name', 'Home Team Goals', 'Away Team Goals', 'Outcome']].head())
```

##### Visualization: Top Winning Teams

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the top winning teams
plt.figure(figsize=(10, 6))
sns.barplot(x=top_winning_teams.values, y=top_winning_teams.index, palette='viridis')
plt.xlabel('Number of Wins')
plt.ylabel('Teams')
plt.title('Top 10 Winning Teams in World Cup History')
plt.show()
```

##### 2. Goals Scored Analysis

```python
# Calculate total goals scored by each team (home and away)
home_goals = matches_df.groupby('Home Team Name')['Home Team Goals'].sum()
away_goals = matches_df.groupby('Away Team Name')['Away Team Goals'].sum()

# Combine the goals scored
total_goals = home_goals.add(away_goals, fill_value=0).sort_values(ascending=False)

# Display the top 10 teams with the most goals
top_scoring_teams = total_goals.head(10)
print(top_scoring_teams)
```

##### Visualization: Top Scoring Teams

```python
# Plot the top scoring teams
plt.figure(figsize=(10, 6))
sns.barplot(x=top_scoring_teams.values, y=top_scoring_teams.index, palette='magma')
plt.xlabel('Total Goals Scored')
plt.ylabel('Teams')
plt.title('Top 10 Scoring Teams in World Cup History')
plt.show()
```

##### 3. Performance by Stage

```python
# Analyze the performance in different stages
stage_wins = matches_df[matches_df['Winner'] != 'Draw'].groupby(['Stage', 'Winner']).size().unstack(fill_value=0)

# Display the stage-wise wins for the top team
top_team_stage_wins = stage_wins.loc[:, stage_wins.sum().idxmax()]
print(top_team_stage_wins)
```

##### Visualization: Performance by Stage

```python
# Plot the performance by stage for the top team
plt.figure(figsize=(10, 6))
top_team_stage_wins.plot(kind='bar', color='c')
plt.xlabel('Stage')
plt.ylabel('Number of Wins')
plt.title('Performance by Stage for the Top Winning Team')
plt.show()
```

##### 4. Impact of Attendance on Performance

```python
# Analyze the impact of attendance on match outcomes
matches_df['Attendance'] = pd.to_numeric(matches_df['Attendance'], errors='coerce')

# Average attendance for wins, losses, and draws
attendance_analysis = matches_df.groupby('Outcome')['Attendance'].mean()
print(attendance_analysis)
```

##### Visualization: Impact of Attendance on Performance

```python
# Plot the average attendance for different outcomes
plt.figure(figsize=(10, 6))
sns.barplot(x=attendance_analysis.index, y=attendance_analysis.values, palette='coolwarm')
plt.xlabel('Match Outcome')
plt.ylabel('Average Attendance')
plt.title('Average Attendance by Match Outcome')
plt.show()
```
